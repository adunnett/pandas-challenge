```python
# Dependencies and Setup
import pandas as pd

import csv
import os

filepath = os.path.join ("Resources", "PyBank.csv")

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df.head()
```

## Player Count

* Display the total number of players


```python
players = df.loc [:, ['SN', 'Age', 'Gender']]
players = players.drop_duplicates()
player_count = players.count()[0]
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(df['Item ID'].unique())
average_price = df['Price'].mean()
number_of_purchases = df['Purchase ID'].count()
total_revenue = sum(df['Price'])
summary_table = pd.DataFrame({'Number of Unique Items' : [unique_items],
                             'Average Price' : [average_price],
                             'Number of Purchases' : [number_of_purchases],
                             'Total Revenue' : [total_revenue]})
summary_table['Average Price'] = summary_table['Average Price'].map('${0:,.2f}'.format)
summary_table['Total Revenue'] = summary_table['Total Revenue'].map('${0:,.2f}'.format)
summary_table.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count = players['Gender'].value_counts()
gender_percent = gender_count/player_count
gender_df = pd.DataFrame({'Percentage of Players' : gender_percent,
             'Total Players' : gender_count})
gender_df['Percentage of Players'] = gender_df['Percentage of Players'].map('{:,.2%}'.format) 
gender_df.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase = df.groupby(['Gender']).count()['Price']
gender_purchase_average = df.groupby(['Gender']).mean()['Price']
gender_purchase_total = df.groupby(['Gender']).sum()['Price']
pp_average_total = gender_purchase_total/gender_count
gender_summary = pd.DataFrame({'Purchase Count' : gender_purchase,
                              'Average Purchase Price' : gender_purchase_average,
                              'Total Purchase Value' : gender_purchase_total,
                              'Avg Total Purchase per Person' : pp_average_total})
gender_summary['Average Purchase Price'] = gender_summary['Average Purchase Price'].map('${0:,.2f}'.format)
gender_summary['Total Purchase Value'] = gender_summary['Total Purchase Value'].map('${0:,.2f}'.format)
gender_summary['Avg Total Purchase per Person'] = gender_summary['Avg Total Purchase per Person'].map('${0:,.2f}'.format)
gender_summary.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
age_df['Age Ranges'] = pd.cut(df['Age'], bins, labels = group_names)
grouped_age = age_df.groupby('Age Ranges')
age_totals = grouped_age["SN"].nunique()
age_percent = (age_totals/player_count)
age_summary = pd.DataFrame({'Total Count' : age_totals,
                           'Percentage of Players' : age_percent})
age_summary['Percentage of Players'] = age_summary['Percentage of Players'].map('{:,.2%}'.format) 
age_summary.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = df["SN"].count()
avg_price_age = df["Price"].mean()
purchase_value = df["Price"].sum()
avg_purchase_pp = purchase_count/age_totals
purchasing_summary = pd.DataFrame({"Purchase Count": purchase_count,
                                 "Average Purchase Price": avg_price_age,
                                 "Total Purchase Value":purchase_value,
                                 "Average Purchase Total per Person": avg_purchase_pp})
purchasing_summary.index.name = None
purchasing_summary.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Average Purchase Total per Person":"${:,.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
total_purchase = df.groupby(['SN']).count()['Price']
purchase_average = df.groupby(['SN']).mean()['Price']
purchase_total = df.groupby(['SN']).sum()['Price']
top_spenders = pd.DataFrame({"Purchase Count": total_purchase,
                          "Average Purchase Price": purchase_average,
                          "Total Purchase Value": purchase_total})
top_spenders.sort_values("Total Purchase Value",ascending=False).round(2).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_id = df.groupby(df['Item ID'])
unique_items = item_id['Item ID'].unique().str[0]
item_name = item_id['Item Name'].unique().str[0]
item_count = item_id['Age'].count()
item_price = item_id['Price'].unique().str[0]
item_total = item_id['Price'].sum()
item_summary =pd.DataFrame({'Item ID':unique_items,
                'Item Name':item_name,
                'Item Price':item_price,
                'Item Count':item_count,
                'Total Purchase':item_total})
item_summary = item_summary.sort_values('Item Count', ascending=False)
item_summary_df = item_summary[['Item Name','Item Count','Item Price','Total Purchase']]
item_summary_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
items_most_profitable = item_summary.sort_values('Total Purchase', ascending=False)
items_most_profitable.head()
```
